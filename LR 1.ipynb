```python
import numpy as np
data_path = "wine.data"
data = np.genfromtxt(data_path, delimiter=";")
print(data)
```

Тип переменной и форма (shape)

```python
print ( "Data type : ", type(data) )
print ( "Data shape : ", data.shape )
print ( data[-4:] )
```

Получение типа набора данных, строки, элемента

```python
data1 = np.genfromtxt(data_path, delimiter=",", dtype=None)
print('Shape of the dataset:', data1.shape)
print('Dataset type:', type(data1))
print('A single row of the dataset is type of:', type(data1[0]))
print('Types of elements:', type(data1[0][1]), type(data1[0][4]))
print('Dataset:')
print(data1)
```

Указание типа столбцов при загрузке данных

```python
dt = np.dtype("f8, f8, f8, f8, U30")
data2 = np.genfromtxt(data_path, delimiter=";", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][4]))
print('Dataset slice:')
print(data2[:10])
```

Построение графиков с использованием Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Данные из отдельных столбцов
fixed_acidity = [] # fixed acidity
volatile_acidity = [] # volatile acidity
citric_acid = [] # citric acid
residual_sugar = [] # residual sugar

# Выполняем обход всей коллекции data2
for dot in data2:
    fixed_acidity.append(dot[0])
    volatile_acidity.append(dot[1])
    citric_acid.append(dot[2])
    residual_sugar.append(dot[3])

# Строим графики по проекциям данных
plt.figure(1)
AbrauDurso, = plt.plot(fixed_acidity [:50], volatile_acidity [:50], 'ro', label=' Abrau-Durso ')
Prosecco, = plt.plot(fixed_acidity [50:75], volatile_acidity [50:75], 'g^', label=' Prosecco ')
Chardonnay, = plt.plot(fixed_acidity [75:100], volatile_acidity [75:100], 'bs', label=' Chardonnay ')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('fixed acidity')
plt.ylabel('volatile acidity')

plt.figure(2)
AbrauDurso, = plt.plot(fixed_acidity [:50], citric_acid [:50], 'ro', label=' Abrau-Durso')
Prosecco, = plt.plot(fixed_acidity [50:75], citric_acid [50:75], 'g^', label=' Prosecco ')
Chardonnay, = plt.plot(fixed_acidity [75:100], citric_acid [75:100], 'bs', label=' Chardonnay')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('fixed acidity')
plt.ylabel('citric acid ')

plt.figure(3)
AbrauDurso, = plt.plot(fixed_acidity [:50], residual_sugar[:50], 'ro', label=' Abrau-Durso')
Prosecco, = plt.plot(fixed_acidity [50:75], residual_sugar[50:75], 'g^', label=' Prosecco')
Chardonnay, = plt.plot(fixed_acidity [75:100], residual_sugar[75:100], 'bs', label=' Chardonnay ')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('fixed acidity')
plt.ylabel('residual sugar')

plt.show()
```
